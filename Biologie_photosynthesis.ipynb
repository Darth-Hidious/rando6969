```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Zeit (min)": [0, 15, 30, 45, 60],
    "CO2-reich (Licht)": [3.43, 4.33, 5.57, 6.64, 7.67],
    "CO2-arm (Licht)": [2.75, 2.82, 3.06, 3.33, 3.73],
    "Dunkelheit": [9.15, 8.97, 8.88, 8.79, 8.70]
}

df = pd.DataFrame(data)

# So this how one can increase the size, smh
plt.figure(figsize=(10, 7), dpi=300)
plt.plot(df["Zeit (min)"], df["CO2-reich (Licht)"], color='blue', linewidth=2.5, label="CO₂-reich (Licht)")
plt.plot(df["Zeit (min)"], df["CO2-arm (Licht)"], color='green', linewidth=2.5, label="CO₂-arm (Licht)")
plt.plot(df["Zeit (min)"], df["Dunkelheit"], color='black', linewidth=2.5, label="Dunkelheit")
plt.xlabel("Zeit (min)", fontsize=14, labelpad=10, fontweight='bold')
plt.ylabel("O₂-Gehalt [mg/l]", fontsize=14, labelpad=10, fontweight='bold')
plt.title("Veränderung des O₂-Gehalts im Wasser durch Egeria densa", fontsize=16, fontstyle='italic', fontweight='bold')
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1), fontsize=12, frameon=False)
plt.tight_layout()
plt.savefig("oxygen_variation_egeria_densa.png", dpi=300, bbox_inches="tight")
plt.show()
#I used chatGPT for this, didn't know how to make the image high quality
```
