```python
import matplotlib.pyplot as plt
import numpy as np

t2 = [0.13675204, 0.17608514, 0.21811235, 0.25981958, 0.30126377, 0.3411]
h2 = [1.4, 1.8, 2.2, 2.6, 3, 3.4]

x_error = [0.00075545, 0.00283917, 0.00057167, 0.00053743, 0.00031926, 0.00035683]
y_error=[0.02]*6

m, b = np.polyfit(t2, h2, 1)
equation = f'y = {m:.4f}x + {b:.4f}'

plt.text(0.155, 3.1, equation, fontsize=12)
plt.grid(True)
plt.errorbar(t2,h2,yerr=y_error, xerr=x_error,fmt="-")
plt.xlabel('t^2 in s2')
plt.ylabel('2h in m')
plt.title('Scatter plot von t2 vs 2h mit error bars')
plt.show()
print(equation)
```

```python
import matplotlib.pyplot as plt
import numpy as np

g = [10.238, 10.222, 10.087, 10.007, 9.958, 9.968]
h2 = [1.4, 1.8, 2.2, 2.6, 3, 3.4]

y_error = [0.1568, 0.20017,0.09543, 0.07971, 0.06722, 0.05955]
x_error=[0.02]*6

m, b = np.polyfit(h2, g, 1)
equation = f'y = {m:.4f}x + {b:.4f}'
p = np.poly1d([m, b])

plt.scatter(h2, g)
plt.plot(h2, p(h2))
plt.grid(True)
plt.errorbar(h2,g,yerr=y_error, xerr=x_error,fmt="-")
plt.xlabel('2h in m')
plt.ylabel('g in m/s^2')
plt.title('Scatter plot von g vs 2h mit error bars')
plt.show()
print(equation)
```
