```python
import numpy as np
import matplotlib.pyplot as plt

def time_period(length, gravity=9.81):
    return 2 * np.pi * np.sqrt(length / gravity)

lengths = np.linspace(0.01, 1, 100)

time_periods = time_period(lengths)

# Plotting the graph of length vs time period
plt.plot(lengths, time_periods)
plt.xlabel('Length (m)')
plt.ylabel('Time period (s)')
plt.title('Dependence of length of the pendulum and the time period')
plt.show()
```
