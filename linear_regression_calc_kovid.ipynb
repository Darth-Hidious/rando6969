```python
import numpy as np
import matplotlib.pyplot as plt

L = np.array([0.40, 0.60, 0.70, 0.80, 0.90, 1.00])
T = np.array([1.29, 1.61, 1.69, 1.81, 1.93, 2.00])


L_mean = np.mean(L)
T_mean = np.mean(T)

b = np.sum((L - L_mean) * (T - T_mean)) / np.sum((L - L_mean)**2)


a = T_mean - b * L_mean


fig, ax = plt.subplots()

# Plot the points
ax.plot(L, T, 'bo', label='Points')

# Plot the best fit line
ax.plot(L, a + b * L, 'r--', label='Best fit line')

ax.set_xlabel('L (m)')
ax.set_ylabel('T (s)')
ax.set_title('Linear regression of T vs L')


ax.legend()
plt.show()

# Print the equation of the best fit line
print(f"The equation of the best fit line is T = {a:.2f} + {b:.2f}L")
```
